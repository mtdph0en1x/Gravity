```python
def get_meaningful_region_of_a_track(points_x, points_y, start_x, finish_x):
    #convert to a numpy array
    points_x = np.array(points_x)
    points_y = np.array(points_y)
    
    #find closest points to the start/finish
    start_distances = np.abs(points_x - start_x)
    finish_distances = np.abs(points_x - finish_x)
    
    #ids of those points 
    start_idx = np.argmin(start_distances)
    finish_idx = np.argmin(finish_distances)
    
    #ensure that we are  going from start to finish
    if start_idx > finish_idx:
        start_idx, finish_idx = finish_idx, start_idx
    #extract the racing segment between start and finish X-coordinates
    racing_x = points_x[start_idx:finish_idx + 1]
    racing_y = points_y[start_idx:finish_idx + 1]
    
    return racing_x, racing_y
```

```python
import numpy as np

def get_track_length(points_x, points_y, start_x, finish_x):
    """
    Calculate track distances using meaningful regions:
    - Straight distance: horizontal span between start_x and finish_x
    - Track length: actual path length between closest X-coordinates
    - Efficiency - ratio between two on the top
    """
    #get meaningful regions of a track
    racing_x, racing_y = get_meaningful_region_of_a_track(
        points_x, points_y, start_x, finish_x
    )
    
    
    #calculate actual track length (sum of segment distances)
    if len(racing_x) > 1:
        dx = np.diff(racing_x)
        dy = np.diff(racing_y)
        track_length = np.sum(np.sqrt(dx**2 + dy**2))
    else:
        track_length = 0
    
    #straight line distance (horizontal span)
    straight_distance = abs(finish_x - start_x)
    
    #efficiency ratio (how simple the track is)
    efficiency = straight_distance / track_length if track_length > 0 else 0
    
    return pd.Series([track_length, straight_distance, efficiency])

df[['actual_track_length', 'straight_distance', 'track_efficiency']] = df.apply(
    lambda row: get_track_length(
        row['points_x_parsed'], 
        row['points_y_parsed'],
        row['start_x'],
        row['finish_x']
    ), axis=1
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_track(points_x, points_y, start_x, start_y, finish_x, finish_y, title):
    '''A quick look into how track looks like'''
    plt.figure(figsize=(12, 6))
    plt.plot(points_x, points_y, 'b-', linewidth=2, label='Track')

    #start/finish coordinates
    plt.scatter(start_x, start_y, color='green', s=100, label='Start', marker='o', edgecolor='black', linewidth=2)
    plt.scatter(finish_x, finish_y, color='red', s=100, label='Finish', marker='s', edgecolor='black', linewidth=2)
    #get meaningful regions of a track
    racing_x, racing_y = get_meaningful_region_of_a_track(
        points_x, points_y, start_x, finish_x
    )
    
    
    plt.scatter(racing_x[0], racing_y[0], color='black', s=100, label='Start', marker='p', edgecolor='white', linewidth=2)
    plt.scatter(racing_x[-1], racing_y[-1], color='white', s=100, label='Finish', marker='P', edgecolor='black', linewidth=2)
    
    plt.title(title)
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.axis('equal')
    plt.show()


def plot_track_from_df(df, index):
    '''Plot track using dataframe row'''
    row = df.iloc[index]
    plot_track(
        points_x=row['points_x_parsed'],
        points_y=row['points_y_parsed'],
        start_x=row['start_x'],
        start_y=row['start_y'],
        finish_x=row['finish_x'],
        finish_y=row['finish_y'],
        title=f"Track: {row['name']} (Level: {row['level']})"
    )

```

```python
def analyze_track_terrain(points_x, points_y,start_x,finish_x):
    """
    Analyze terrain with total climb/drop distances
    """
    #get meaningful regions of a track
    racing_x, racing_y = get_meaningful_region_of_a_track(
        points_x, points_y, start_x, finish_x
    )
    
    
    if len(racing_x) < 3:
        return pd.Series([0] * 8)
        
    dx=np.diff(racing_x)
    dy=np.diff(racing_y)
    #calculate segment distances
    segment_distances = np.sqrt(dx**2 + dy**2)
    
    #basic counts between each two points
    climbs = np.sum(dy > 0)
    drops = np.sum(dy < 0)
    flat_sections = np.sum(dy == 0)
    
    climb_distance = np.sum(segment_distances[dy > 0])  # total distance while climbing
    drop_distance = np.sum(segment_distances[dy < 0])   # total distance while dropping
    flat_distance = np.sum(segment_distances[dy == 0])  # total distance on flat
    
    # significant changes in steepness counts
    steep_climbs = np.sum(dy > 50)
    steep_drops = np.sum(dy < -50)

    y_variance = np.var(racing_y)
    elevation_range= np.max(racing_y) - np.min(racing_y)
    
    return pd.Series([
        climbs, drops, flat_sections,
        climb_distance, drop_distance, flat_distance,
        steep_climbs, steep_drops,y_variance,elevation_range
    ])

terrain_features = [
    'climbs', 'drops', 'flat_sections',
    'climb_distance', 'drop_distance', 'flat_distance',
    'steep_climbs', 'steep_drops','y_variance','elevation_range'
]

df[terrain_features] = df.apply(
    lambda row: analyze_track_terrain(row['points_x_parsed'], row['points_y_parsed'],row['start_x'],row['finish_x']), 
    axis=1
)
```

```python
def count_direction_changes(points_x, points_y, start_x, finish_x, threshold=60):
    """Count significant direction changes in track"""
    if len(points_x) < 3:
        return 0
        
    racing_x, racing_y = get_meaningful_region_of_a_track(
        points_x, points_y, start_x, finish_x
    )
    
    changes = 0
    for i in range(2, len(racing_x)):
        # Calculate vectors
        v1 = (racing_x[i-1] - racing_x[i-2], racing_y[i-1] - racing_y[i-2])
        v2 = (racing_x[i] - racing_x[i-1], racing_y[i] - racing_y[i-1])
        
        # Calculate angle change
        angle_change = np.arccos(np.clip(
            np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)), -1, 1
        )) * 180 / np.pi
        
        if angle_change > threshold:
            changes += 1
    
    return changes

df['direction_changes']=df.apply(
    lambda row: count_direction_changes(row['points_x_parsed'], row['points_y_parsed'],row['start_x'],row['finish_x']), 
    axis=1
)
```

```python
def remove_spikes(points_x,points_y,start_x,finish_x):
    
    x=np.array(points_x)
    y=np.array(points_y)
    if len(points_x) < 3:
        return pd.Series([0] * 2)
        
    dx = np.diff(x) 
    dy = np.diff(y) 
    
    removed_idx = []
    
    for i in range(0,len(dx)-1):
        if (dx[i]+dx[i+1])*5<(dy[i]+dy[i+1]):  # ((x[i+1] - x[i]) + (x[i+2] - x[i+1])) * threshold < ((y[i+1] - y[i]) + (y[i+2] - y[i+1]))
            removed_idx.append(i)

            
    keep_idx = [i for i in range(len(x)) if i not in removed_idx]
    cleaned_x = x[keep_idx]
    cleaned_y = y[keep_idx]

    return pd.Series([cleaned_x, cleaned_y])

clean_track=['cleaned_x','cleaned_y']
df[clean_track]=df.apply(
    lambda row: remove_spikes(row['points_x_parsed'], row['points_y_parsed'],row['start_x'],row['finish_x']), 
    axis=1
)
            
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df[features].fillna(0)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
#Check PC1 
print("\nPC1 loadings:")
loadings = pca.components_[0]
for i, loading in enumerate(loadings):
    print(f"  {features[i]}: {loading:.3f}")
```

```python
# Plot explained variance
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), 
         pca.explained_variance_ratio_, 'bo-')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Variance Explained by Each PC')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), 
         np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Variance Explained')
plt.axhline(y=0.95, color='k', linestyle='--', alpha=0.5, label='95%')
plt.axhline(y=0.80, color='g', linestyle='--', alpha=0.5, label='80%')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(X_scaled)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_scaled, kmeans.labels_, metric='euclidean')
```

```python
K = range(5, 15)
fits = []
score = []


for k in K:
    # train the model for current value of k on training data
    model = KMeans(n_clusters = k, random_state = 0, n_init='auto').fit(X_pca)
    
    # append the model to fits
    fits.append(model)
    
    # Append the silhouette score to scores
    score.append(silhouette_score(X_scaled, model.labels_, metric='euclidean'))
```
